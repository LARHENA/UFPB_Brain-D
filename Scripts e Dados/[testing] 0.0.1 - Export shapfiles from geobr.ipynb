```python
# Load the Brazilian states shapefile (2020)
brazil_states = geobr.read_state(year=2020)

# Convert to WGS84 (EPSG:4326)
brazil_states_wgs84 = brazil_states.to_crs(epsg=4326)

# Export to Shapefile
output_path = './shp/brazil_states_2020.shp'
brazil_states_wgs84.to_file(output_path, driver='ESRI Shapefile', encoding='utf-8')
```
