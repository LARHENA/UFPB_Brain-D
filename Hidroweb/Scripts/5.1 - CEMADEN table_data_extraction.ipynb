```python
all_files = glob.glob(cemaden_path + r'\1 - Organized data\*.h5')
# all_files = all_files[:5]
all_files
```

```python
# list of 270 items 
original_list = all_files

# Divide the list into ten lists of 27 items each
divided_lists = [original_list[i:i+27] for i in range(0, len(original_list), 27)]
final_list = [[]*10]
print(f"final_list: {final_list}")
# Print the result
for i, sublist in enumerate(divided_lists):
    if i == 0:
        final_list = [sublist]
        print(f"List {i}: {final_list}")
    else:
        # print(f"List {i + 1}: {sublist}")
        final_list.append(sublist)
        print(f"List {i}: {final_list}")

print(f"final_list conclu√≠da: {final_list}")
print(f"final_list length: {len(final_list)}")
```

```python
for i in range(10):
    count = 0
    len_files = len(final_list[i])
    for filename in final_list[i]:
        df = pd.read_hdf(filename, 'table_data')
        df = df[['gauge_code', 'datetime', 'rain_mm']]
        if count == 0:
            df_total = df.copy(deep = True)
        else:
            df_total = pd.concat([df_total, df], ignore_index = True)
        del df
        count = count + 1
        print("File", count, "/", len_files, "|", filename)
    print(count, "files are loaded")
    df_total.to_hdf(general_path + '/Consolidated Files/CEMADEN/BRASIL_CEMADEN_TOTAL_'+str(i)+'.h5', 'table_data', mode = 'w', append = False, complevel = 9, encoding="cp860")
```

```python
all_files = glob.glob(general_path + r'\Consolidated Files\CEMADEN\*.h5')
all_files
```

```python
count = 0
len_files = len(all_files)
for filename in all_files[:]:
    df = pd.read_hdf(filename, 'table_data')
    df = df[['gauge_code', 'datetime', 'rain_mm']]
    df['date'] = df['datetime'].dt.date
    df['date'] = pd.to_datetime(df['date'])
    df = df[['gauge_code', 'rain_mm', 'date']]
    df = df.groupby(['gauge_code', 'date'])['rain_mm'].sum().reset_index()
    if count == 0:
        df_total = df.copy(deep = True)
    else:
        df_total = pd.concat([df_total, df], ignore_index = True)
    count = count + 1
    print("File", count, "/", len_files, "|", filename)
print(count, "files are loaded")
```
