### Reading Raw General File

```python
df_total_hdf = pd.read_hdf(general_path+'/Consolidated Files/BRASIL_RAW_NOT_CEMADEN.h5')
```

```python
df_total_hdf
```

```python
df_total_hdf = df_total_hdf.drop_duplicates(ignore_index = True)
df_total_hdf = df_total_hdf.reset_index(drop = True)
df_total_hdf['Year'] = pd.DatetimeIndex(df_total_hdf['Date']).year
df_total_hdf
```

```python
df_total_hdf.dtypes
```

```python
df_total_hdf['Date'].max()
```

```python
df_total_hdf_2022 = df_total_hdf[(df_total_hdf['Year'] < 2023)]
df_total_hdf_2022 = df_total_hdf_2022.reset_index(drop = True)
df_total_hdf_2022
```

```python
df_total_hdf_2022.to_hdf(general_path + '/Consolidated Files/BRASIL_RAW_1855_2022.h5', 'table_data', mode = 'w', append = False, complevel = 9, encoding="cp860")
```
