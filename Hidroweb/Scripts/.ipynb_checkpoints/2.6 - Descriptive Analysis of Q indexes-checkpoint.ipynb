```python
df_q = pd.read_hdf(general_path + '/Quality/BRASIL_Q_QUALITY_INDEX_by_month.h5')
df_q
```

```python
agg_functions = {
    'Availability': 'mean',
    'Gaps': 'mean',
    'Weekday': 'mean',
    'Outlier': 'mean',
    'Quality Index': 'mean'
}

# Create pivot table
df_date_q = pd.pivot_table(df_q, values=['Availability', 'Gaps', 'Weekday', 'Outlier', 'Quality Index'],
                           index='Date', aggfunc=agg_functions)
df_date_q.reset_index(inplace=True, drop=False)
df_date_q
```

```python
date_column = df_date_q['Date']
list_dates = pd.to_datetime(date_column).tolist()
xticks_labels = [dt.strftime('%Y-%m-%d') for dt in list_dates]
len(xticks_labels)
```

```python
df_date_q['Year'] = pd.to_datetime(df_date_q['Date']).dt.year
list_years = pd.to_datetime(df_date_q['Year']).values.tolist()
len(list_years)
```

```python
xticks_labels = len(list_years)*[""]
count = 60
for i in range(len(list_years)):
    if (str(list_years[i])[-1:] == "0" and str(list_years[i-1])[-1:] == "9") or i == 0:
        print(str(list_years[i]), str(list_years[i-1]))
        xticks_labels[i] = str(list_years[i])
    elif (str(list_years[i])[-1:] == "5" and str(list_years[i-1])[-1:] == "4") or i == 0:
        print(str(list_years[i]), str(list_years[i-1]))
        xticks_labels[i] = str(list_years[i])
    else:
        xticks_labels[i]  = ""
print(xticks_labels)
```

```python
df_state_q = pd.pivot_table(df_q, values = ['Availability','Gaps', 'Weekday', 'Outlier', 'Quality Index'], index=['State'], aggfunc = {
    'Availability': 'mean',
    'Gaps': 'mean',
    'Weekday': 'mean',
    'Outlier': 'mean',
    'Quality Index': 'mean'
})
df_state_q.reset_index(inplace=True, drop=False)
df_state_q.sort_values("Quality Index", inplace=True, ignore_index=True)
df_state_q
```

```python
labels = df_state_q['Quality Index'].round(2).values.tolist()
labels
```

```python
fig, ax = plt.subplots(figsize = (16, 6))

ax.bar(df_state_q['State'], df_state_q['Quality Index'], color='grey', edgecolor = 'black')
ax.set_ylim(min(labels)*0.95, max(labels)*1.05)
tnr = {'fontname':'Times New Roman', 'fontsize':16}

# ax.set_ylabel("Quantidade de Estações", size = 14, **tnr)
# ax.set_title("", size=16, **{'fontname':'Times New Roman'})
# ax.set_xticks(xticks)
# ax.set_xticklabels(xticks_labels)

# Make some labels.
rects = ax.patches
  
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height+0.5, label,
            ha='center', va='bottom', rotation = 90, **tnr)

plt.setp(plt.gca().get_xticklabels(), rotation = 90, **tnr)
plt.setp(plt.gca().get_yticklabels(), **tnr)
plt.tight_layout() 
filename = general_path + '/Figures/average_quality_index_by_state_by_month.jpeg'
print(filename)
plt.savefig(filename, format='jpeg', dpi=1200, bbox_inches='tight') 
plt.show()
plt.close()
```
